# 13. Splitting the data into Features and Targets

```python
X = new_data.drop(columns='Class',axis=1)
Y = new_data['Class']
```

```python
print(X)
```

```python
print(Y)
```

# 14. Splitting the data into Training data and Testing data

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

# 15. Model Training using logistic regression
   1. Logistic regression is used for binary classification problems
   2. Predicting the categorical dependent variable using a given set of independent variables

```python
model = LogisticRegression()
```

```python
# Training thelogistic regression model with training data
model.fit(X_train, Y_train)
```

# 16. Model Evaluation

```python
# Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on training data is : ", training_data_accuracy)
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy= accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on test data is : ", test_data_accuracy)
```
